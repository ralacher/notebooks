# Monte Carlo

Imports required for the program

```python
from __future__ import division
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import nbconvert
```

Constants used for the program.


Each iteration of the Monte Carlo simulation (**NUM_ITERATIONS**) will run five times, with 10000 data points (**NUM_ROUNDS**) per iteration. **LEN_SIDE** stores the side length of the square.

```python
# Define constants
NUM_ITERATIONS = 5
NUM_ROUNDS = 100
LEN_SIDE = 1.5
```

The function that the Monte Carlo simulation will use:

$$ x^2 + y^2 + xy = 1 $$

```python
# Define function
def f(x,y):
    result = x**2 + y**2 + x * y
    if result <= 1:
        return True
    return False
```

The driver of the program to run the Monte Carlo simulation.

```python
%matplotlib inline

# results will store the number of points within the ellipsis each iteration
results = []

# Outer loop runs NUM_ITERATIONS times
for i in xrange(NUM_ITERATIONS):
    # Plotting
    x = np.linspace(-1.5, 1.5, 100)
    y = np.linspace(-1.5, 1.5, 100)
    X, Y = np.meshgrid(x,y)
    radius = X**2 + Y**2 + X * Y
    plt.contour(X,Y,radius,[1])
    plt.title('Monte Carlo Simulation')
    # End Plotting

    
    # num_in stores the number of points within the ellipsis for this round
    num_in = 0
    
    # Inner loop
    for j in xrange(NUM_ROUNDS):
        #  Uniform distribution [-1.5, 1.5]
        x = np.random.uniform(low=-1.5,high=1.5)
        y = np.random.uniform(low=-1.5,high=1.5)
    
        # Check if f(x,y) is within the ellipsis
        # If so, plot the point and increment the number of points within
        # the circle
        if f(x, y):
            num_in += 1
            plt.plot(x, y,'ro')
        else:
            plt.plot(x, y, 'go')
    
    # Append the number of points within the circle to the array
    results.append(num_in)
    
    # Print the results for this round
    print 'Area for round %s= %s' % (i, ((1 - (num_in / NUM_ROUNDS)) * LEN_SIDE * 4))
    
# Print the overall results
print 'Area final= %s' % ((1 - sum(results) / (NUM_ITERATIONS * NUM_ROUNDS)) * LEN_SIDE * 4)
```
