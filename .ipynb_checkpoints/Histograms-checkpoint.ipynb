```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x = np.random.randn(1000000)
plt.hist(x,100)
np.std(x)
```

```python
%matplotlib inline
y = np.random.uniform(low=0, high=1, size=1000000)
plt.hist(y, 50)
np.std(y)
```
