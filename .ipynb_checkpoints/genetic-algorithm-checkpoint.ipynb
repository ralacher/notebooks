The purpose of this assignment was to research and implement a genetic algorithm. The requirements of the assignment were that each **bitstring** must be of length 10 and each **population** must consist of 10 candidates. **Mutation chance** and the **fitness function** were chosen arbitrarily.

The program uses bitstrings (strings of 0s and 1s, e.g. 011010) to model the problem. The goal is to maximize the **objective function** by "evolving" the population of bitstrings over 500 generations. 

To accomplish this evolution, 10 random bitstrings are generated at the start of the program. These bitstring candidates will be evaluated against the fitness function -- "fit" candidates will be of greater value after going through the fitness function, and "less fit" candidates will be of lower value.

The meat of the algorithm lies in the **selection**, **crossover**, and **mutation** portion of the algorithm, which takes two parent bitstrings and produces a single child bitstring. 

In the **selection** stage, a random index will be generated and used to partition one of the parent bitstrings. For example, a bitstring $$0000011111$$ partitioned at the fifth element would return a left-hand side of $$00000$$ and a right-hand side of $$11111$$. 

The program will then **crossover** the two parents based upon the index value. The child bitstring will consist of the left-hand side of parent one's bitstring and the right-hand side of parent two's bitstring, resulting in an "evolved" child. Using the above bitstring as parent one, $$1111100000$$ as parent two, and an index of five, the child bitstring would be $$0000000000$$.

The child bitstring then has a chance to **mutate** and have one of its bits modified. The child has a $$1 \over{MUTATION\ CHANCE}$$ chance to be mutated. If this occurs, then one of the child's bits will be randomly selected and swapped. Using the above example, the child bitstring $$0000000000$$ with a random index of five would be mutated to $$0000100000$$.

This simulation is ran for 500 generations. "Fit" parent bitstrings are selected each generation and produce child bitstrings, which become the parent population in subsequent generations. As the simulation is ran the average fitness of the population is recorded, as well as the number of mutations, and visualized using the **matplotlib** graphing library.

```python
from __future__ import division
from random import randint
from math import ceil, e
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

The program first starts with the imports. **numpy** and **matplotlib** were utilized to visualize the output of the program.

```python
BITSTRING_LEN = 7
POPULATION_LEN = 10
MUTATION_CHANCE = 999
MAXIMUM_VALUE = 1023
EPOCHS = 1000
```

The program then defines the constants.

```python
def fitness_function(value):
    ''' Fitness function to be maximized by the genetic algorithm
    @param value: integer value to be maximized
    @return: integer result of objective function
    '''
    return -((value - 4) ** 4 - (value - 4) ** 3 - 15 * (value - 4) ** 2 + 9 * (value - 4))
```

The fitness function I chose was $$f(x)=-(x^4 - x^3 - 15x^2 + 9x).$$ 

```python
def bitstring_to_int(value):
    ''' Function to generate a base 10 integer from a bitstring
    @param value: bitstring
    @return: integer value of bitstring
    '''
    out = 0
    for i, v in enumerate(reversed(value)):
        out += int(v) * (2 << i) / 2
    return out
```

A helper function to convert the bitstring to a base-10 integer value.

```python
def create_permutation():
    ''' Utility function to generate a random bitstring
    @return: bitstring of length BITSTRING_LEN
    '''
    return ''.join(map(str, [randint(0, 1) for i in xrange(0, BITSTRING_LEN)]))
```

A helper function to generate a random bitstring of length **BITSTRING_LEN**.

```python
def mutate(parent_one, parent_two):
    ''' Function to select, crossover, and mutate values
    @param parent_one: parent bitstring
    @param parent_two: parent bitstring
    @param mutation_chance: chance for a mutation to occur
    @return: child bitstring
    '''
    # Random index where partition will occur
    index = randint(0, BITSTRING_LEN - 1)
    # Crossover
    child = parent_one[0:index]
    child += parent_two[index:]
    # Boolean value to store whether mutation occurs
    mutation = False
    # Chance to mutate element of bitstring
    if randint(0, MUTATION_CHANCE) == 0:
        child = list(child)
        child[randint(0, len(child) - 1)] = str(randint(0, 1))
        child = ''.join(child)
        mutation = True
    return child, mutation
```

The mutation function, which handles the selection, crossover, and mutation portion of the algorithm.

```python
def random_selection(population):
    ''' Function to randomly select two parent bitstrings from the population
    @param population: list of bitstrings
    @return: two selected parent bitstrings
    '''
    choices = []
    max_bitstring = None
    max_value = 0
    for bitstring in population:
        if fitness_function(bitstring_to_int(bitstring)) > max_value:
            max_bitstring = bitstring
        choices.append(bitstring)
    
    for i in xrange(0,20):
        choices.append(max_bitstring)
        
    one = randint(0, len(population) - 1)
    two = randint(0, len(population) - 1)
    
    parent_one = choices[one]
    parent_two = choices[two]
    
    return parent_one, parent_two
```

The random selection function used to randomly select two parent bitstrings to produce a child. The fitnesses of the population are summed and a probability distribution is created. Two random bitstrings are then selected from that distribution and returned.

```python
def simulation():
    ''' Function to run the genetic algorithm '''
    # Lists to store information regarding epochs
    population_average = []
    mutation_generation = []
    
    # Generate 10 random bitstrings to initialize the population
    population = [create_permutation() for i in xrange(0, POPULATION_LEN)]
    
    # Iterate for 500 generations
    for i in xrange(0, EPOCHS):
        new_population = []
        
        # Iterate over population
        for j in xrange(0, len(population)):
            # Randomly select two parent bitstrings from the population
            parent_one, parent_two = random_selection(population)
            # Crossover the parents and return a child bitstring
            child, mutation = mutate(parent_one, parent_two)
            new_population.append(child)
        
        # Calculate average fitness of this generation
        population = new_population
        #average = max([bitstring_to_int(bitstring) for bitstring in population])
        average = sum([bitstring_to_int(c) for c in population]) / len(population)
        #average = average / MAXIMUM_VALUE * 100
        population_average.append(average)
        
        # Record whether a mutation occurred this generation
        if mutation:
            mutation_generation.append(average)
        else:
            mutation_generation.append(None)
    
    return mutation_generation, population_average
```

The simulation function drives the entire algorithm. Lists are used to store the average fitness and whether or not a mutation occurred in that generation. The scenario is ran for 500 generations, at which point the two lists are returned.

```python
def graph(mutations, averages):
    # Calculate number of mutations
    num_mutations = sum([1 for i in mutations if i is not None])
    
    # Graph labels
    plt.xlabel('Generation')
    plt.ylabel('Population Average Fitness %')
    plt.title('Genetic Algorithm')
    # Plot points
    mutations, = plt.plot(mutations, 'ro', label='Mutations (%s)' % num_mutations)
    average, = plt.plot(averages, 'b-', label='Average Fitness')
    mean, = plt.plot([np.mean(averages) for i in averages], 'r-', label='Mean %s' % np.mean(averages))
    
    # Plot legend
    plt.legend(handles=[mutations, average, mean], loc='upper right')
    #plt.legend(handles=[mutations, average], loc='upper right')
    plt.show()
```

The graph function then produced a **matplotlib** graph of the average fitness, using a red dot to indicate generations in which a mutation occurred. 

```python
%matplotlib inline
mutations, averages = simulation()
graph(mutations, averages)
```

From the results, the average fitness rapidly increases and stabilizes at roughly 91% of the maximum value. By increasing the mutation chance from $$1 \over{1000}$$ to $$1 \over{100}$$, the effects of the mutation chance may be observed.

```python
MUTATION_CHANCE = 100
%matplotlib inline
mutations, averages = simulation()
graph(mutations, averages)
```

The higher chance of mutation results in more variations, and more chances for the algorithm to approach the maximum value. The graph above shows the population tending towards roughly 100%.

```python
MUTATION_CHANCE = 10
%matplotlib inline
mutations, averages = simulation()
graph(mutations, averages)
```

However, increasing the mutation chance to $$1 \over{10}$$ may mean that the algorithm fails to converge -- too frequent mutations result in wild variations within the population. It can also be seen that the algorithm, at times, approaches the theoretical maximum.

It's worth noting that genetic algorithms are not guaranteed to converge. In multiple executions of the program, wildly varying results may be seen. 

```python
MUTATION_CHANCE = 5
%matplotlib inline
mutations, averages = simulation()
graph(mutations, averages)
```
